```python
# Transponer b para que tenga la misma estructura que a (2, 3, 5)
c = b.transpose(1, 2, 0)  # o también puedes usar: c = np.transpose(b, (1, 2, 0))

print(f"Forma original de b: {b.shape}")
# b es (5, 2, 3)
print(f"Forma de a (objetivo): {a.shape}")
# a es (2, 3, 5)

# Verificar resultado
print(f"\nForma de c después de transponer: {c.shape}")
print(f"¿c tiene forma (2, 3, 5)? {c.shape == (2, 3, 5)}")

# Mostrar algunos valores para verificar
print("\nPrimera matriz de b (antes de transponer):")
print(b[0, :, :])  # Primera matriz de b

print("\nPrimera matriz de c (después de transponer):")
print(c[0, :, :])  # Primera matriz de c
```
